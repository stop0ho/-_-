# 그룹 애너그램
문자열 배열을 받아 애너그램 단위로 그룹핑하라.
https://leetcode.com/problems/group-anagrams/


애너그램은 문자열 재배열하여 다른 뜻을 가진 단어로 바꾸는 것을 말한다.
=> 문자를 재배열 했을 때 같다 == 문자를 정렬했을 때 동일하다.

## 문자열 정렬
Python의 sorted() 함수는 요소 값으로 반복 가능한 객체를 정렬하는 데 사용되며, 결과값을 리스트로 반환한다

# 내 코드 

```python
def groupAnagrams(strs: list[str]) -> list[list[str]]:
    d = {''.join(sorted(word)) : [] for word in strs}
    for word in strs:
        d[''.join(sorted(word))].append(word)
    result = [d[l] for l in d.keys()]
    return result
```

```python
print(groupAnagrams(["eat","tea","tan","ate","nat","bat"]))
```

```python
print(groupAnagrams([""]))
```

```python
print(groupAnagrams(["a"]))
```

# 추가 공부 : 파이썬의 정렬 함수

```python
# 1. sorted 함수
a = [2, 5, 1, 9, 7]
sorted(a)
```

```python
b = 'zbdaf'
sorted(b)
```

```python
# 정렬된 문자열 리스트를 다시 문자열로 결합
"".join(sorted(b))
```

```python
# sort() 메소드 : 리스트 자체를 정렬(제자리 정렬)
a = [2, 5, 1, 9, 7]
print("정렬 전 : ", a)
a.sort()
print("정렬 후 : ", a)
```

```python
# sorted에 키 지정
c = ['ccc', 'aaaa', 'd', 'bb', 'cc']
sorted(c, key = len)
```

```python
# key에 람다함수
a = ['cde', 'cfc', 'abc']
sorted(a, key=lambda s: (s[0], s[-1]))
```
